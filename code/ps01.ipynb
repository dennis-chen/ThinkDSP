5: Synthesize a wave by creating a spectrum with arbitrary harmonics. 

We start with a wave with 440 Hz, 660 Hz, and 880 Hz frequency components that have the same amplitude.

```python
import cmath
hs = np.empty((5000,),dtype=complex)
hs.fill(cmath.rect(0,0))
hs[440] = cmath.rect(1000,0)
hs[660] = cmath.rect(1000,0)
hs[880] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs,framerate=10000)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

6: Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal.

```python
class SawtoothSignal(thinkdsp.Sinusoid):
    
    def evaluate(self,ts):
        cycles = self.freq * ts + self.offset / (2*pi)
        frac,_ = np.modf(cycles)
        ys = 2*thinkdsp.unbias(frac)
        return ys

sawtooth_signal = SawtoothSignal(freq = 440)
sawtooth_wave = sawtooth_signal.make_wave(duration=0.025,start=0,framerate=11025)
sawtooth_wave.plot()
plt.show()
sawtooth_spectrum = sawtooth_wave.make_spectrum()
sawtooth_spectrum.plot()
plt.show()
```

7: Compute the spectrum of a sawtooth wave. How does the harmonic structure compare to triangle and square waves?

```python
triangle_signal = thinkdsp.TriangleSignal(freq = 440)
triangle_wave = triangle_signal.make_wave(duration=0.025,start=0,framerate=11025)
triangle_spectrum = triangle_wave.make_spectrum()
triangle_spectrum.plot()
plt.show()
square_signal = thinkdsp.SquareSignal(freq = 440)
square_wave = square_signal.make_wave(duration=0.025,start=0,framerate=11025)
square_spectrum = square_wave.make_spectrum()
square_spectrum.plot()
plt.show()
```

The sawtooth wave has harmonics at frequencies that are both even and odd multiples of the fundamental, while the triangle and square waves have harmonics that are odd multiples only. The harmonics of the sawtooth wave also drop off more slowly than the harmonics of the triangle wave.

9) Compute the spectrum of an 1100 Hz square wave sampled at 10 kHz, and compare it to the spectrum of a triangle wave at 1100 Hz sampled at 10 KHz.

```python
square_signal = thinkdsp.SquareSignal(freq = 1100)
square_wave = square_signal.make_wave(duration=0.5,start=0,framerate=10000)
square_spectrum = square_wave.make_spectrum()
square_spectrum.plot()
plt.show()
triangle_signal = thinkdsp.TriangleSignal(freq = 1100)
triangle_wave = triangle_signal.make_wave(duration=0.5,start=0,framerate=10000)
triangle_spectrum = triangle_wave.make_spectrum()
triangle_spectrum.plot()
plt.show()
```

The spectrum of the square wave has more aliased frequencies with higher amplitudes than the aliased frequencies of the triangle wave. The reason the square wave spectrum has aliased frequencies with higher amplitudes is that the square wave's harmonics drop off proportionally to 1/f but the triangle wave's harmonics drop off proportionally to 1/f^2. So, the odd harmonics that fold back due to aliasing are larger in the square wave spectrum. There are also more frequencies that are folded back in the square wave spectrum because square waves have more frequency components than triangle waves.  

10: The triangle and square waves have odd harmonics only; the sawtooth wave has both even and odd harmonics. The harmonics of the square and sawtooth waves drop off in proportion to 1/f; the harmonics of the triangle wave drop off like 1/f2. Can you find a waveform that has even and odd harmonics that drop off like 1/f^2?

```python
hs = np.empty((5000,),dtype=complex)
hs.fill(cmath.rect(0,0))
for freq in range(100,5000,100):
    hs[freq] = 1/(freq**2)
my_spec = thinkdsp.Spectrum(hs,framerate=10000)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

I created a wave above with both even and odd harmonics that drops off propotionally to 1/f^2. Interestingly enough, it reminds me of the noise you briefly hear when plugging or unplugging guitar cables into amplifiers.
