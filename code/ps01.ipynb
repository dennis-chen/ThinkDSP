We'll try creating our own spectrum with arbitrary harmonics now. We start with a pure sinusoid at 440 Hz.

```python
import cmath
hs = np.empty((5000,),dtype=complex)
hs.fill(cmath.rect(0,0))
hs[440] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs,framerate=11056)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

Now we throw in frequencies at 660 Hz and 880 Hz. Our percieved pitch of the sound gets lower! Mixing in higher frequencies causes us to percieve a pitch that corresponds to having a fundamental frequency of 220 Hz, like in the saxaphone.py example. Nifty. 

```python
hs[660] = cmath.rect(1000,0)
hs[880] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs,framerate=11056)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

Here we throw in a wave with only a fundamental frequency of 220 Hz for comparison with the wave above.

```python
hs_220 = np.empty((5000,),dtype=complex)
hs_220.fill(cmath.rect(0,0))
hs_220[220] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs_220,framerate=11056)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

Now we add some frequency components to the 440,660,880 Hz spectrum that are not multiples of the fundamental.

```python
hs[700] = cmath.rect(1000,0)
hs[1000] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs,framerate=11056)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

Sounds a lot worse. And if we add a lot more frequency components that aren't multiples of the fundamental...

```python
hs[600] = cmath.rect(1000,0)
hs[650] = cmath.rect(1000,0)
hs[800] = cmath.rect(1000,0)
hs[900] = cmath.rect(1000,0)
my_spec = thinkdsp.Spectrum(hs,framerate=11056)
my_spec.plot()
plt.show()
my_spec.make_wave().make_audio()
```

It starts to sound less like a clear note and more like noise!

Writing the sawtooth signal class. Used the sinusoid class from thinkdsp as a template, this code is not original!

```python
def sawtooth_func(x):
    y = []
    for val in x:
        remainder = val%(2*pi)
        if remainder < pi:
            y.append(-1 + 2*remainder/pi)
        else:
            y.append(1 - 2*(remainder-pi)/pi)
    return np.asarray(y)

class Sawtooth(thinkdsp.Signal):
    """Represents a sawtooth signal."""
    
    def __init__(self, freq=440, amp=1.0, offset=0, func=sawtooth_func):
        """Initializes a sinusoidal signal. 
        freq: float frequency in Hz
        amp: float amplitude, 1.0 is nominal max
        offset: float phase offset in radians
        func: function that maps phase to amplitude
        """
        self.freq = freq
        self.amp = amp
        self.offset = offset
        self.func = func

    @property
    def period(self):
        """Period of the signal in seconds.
        returns: float seconds
        """
        return 1.0 / self.freq
        
    def evaluate(self, ts):
        """Evaluates the signal at the given times.
        ts: float array of times
        returns: float wave array
        """
        phases = 2*pi * self.freq * ts + self.offset
        ys = self.amp * self.func(phases)
        return ys
    
def SawtoothSignal(freq = 440,offset=0):
    """returns a sawtooth signal"""
    return Sawtooth(freq=freq,amp=1.0,offset=offset,func=sawtooth_func)
    
sawtooth_signal = SawtoothSignal(freq = 440)
sawtooth_wave = sawtooth_signal.make_wave(duration=0.025,start=0,framerate=11025)
sawtooth_wave.plot()
plt.show()
sawtooth_spectrum = sawtooth_wave.make_spectrum()
sawtooth_spectrum.plot()
plt.show()
```

Let's see how the sawtooth spectrum compares to square wave spectrums and triangle wave spectrums.

```python
triangle_signal = thinkdsp.TriangleSignal(freq = 440)
triangle_wave = triangle_signal.make_wave(duration=0.025,start=0,framerate=11025)
triangle_spectrum = triangle_wave.make_spectrum()
triangle_spectrum.plot()
plt.show()
square_signal = thinkdsp.SquareSignal(freq = 440)
square_wave = square_signal.make_wave(duration=0.025,start=0,framerate=11025)
square_spectrum = square_wave.make_spectrum()
square_spectrum.plot()
plt.show()
```

The sawtooth and triangle wave spectrums look the same. If there is a difference, I can't tell that there is one by glancing at the spectrums. The square wave is significantly different, with lots of different frequencies mixed in along with the harmonic frequencies. 

```python
triangle_signal = thinkdsp.TriangleSignal(freq = 1100)
triangle_wave = triangle_signal.make_wave(duration=0.5,start=0,framerate=10000)
triangle_wave.make_audio()
```

```python
triangle_signal = thinkdsp.TriangleSignal(freq = 1100)
triangle_wave = triangle_signal.make_wave(duration=0.5,start=0,framerate=50000)
triangle_wave.make_audio()
```

I can hear the aliased harmonic as a lower pitch! It goes away when the sampling rate increases, cool! Now we can compare square wave and triangle wave spectrums when we have 1100 Hz signals that are sampled at 10 KHz.

```python
triangle_signal = thinkdsp.TriangleSignal(freq = 1100)
triangle_wave = triangle_signal.make_wave(duration=0.5,start=0,framerate=10000)
triangle_spectrum = triangle_wave.make_spectrum()
triangle_spectrum.plot()
plt.show()
square_signal = thinkdsp.SquareSignal(freq = 1100)
square_wave = square_signal.make_wave(duration=0.5,start=0,framerate=10000)
square_spectrum = square_wave.make_spectrum()
square_spectrum.plot()
plt.show()
```
