1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
def make_idft_matrix_0_N(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    return M

def make_idft_matrix_N_2N(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N,2*N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    return M

def make_dft_matrix_0_N(ys):
    return make_idft_matrix_0_N(ys).conj().transpose()

def make_dft_matrix_N_2N(ys):
    return make_idft_matrix_N_2N(ys).conj().transpose()

wave = thinkdsp.TriangleSignal(freq=1).make_wave(duration=1, framerate=8)
idft1 = make_idft_matrix_0_N(wave.ys)
idft2 =  make_idft_matrix_N_2N(wave.ys)
print numpy.allclose(idft1,idft2)
```

It looks like going from N to 2N-1, you get the same idft matrix as going from 0 to N! Numpy confirms that the values in these two matrices are very close to one another. What's the cause of this? The DFT is periodic in the frequency domain! The Nth frequency and the 0th frequency are aliases of each other and so on.

3) So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
signal = thinkdsp.SquareSignal(freq=440)
wave = signal.make_wave(duration=1, framerate=44100)
window = numpy.ones(11)
window /= sum(window)
spectrum = wave.make_spectrum()
ys = numpy.convolve(wave.ys, window, mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
spectrum2 = smooth.make_spectrum()
amps = spectrum.amps
amps2 = spectrum2.amps
ratio = amps2 / amps    
ratio[amps<560] = 0
thinkplot.plot(ratio)

thinkplot.config(xlabel='frequency (Hz)',
                     ylabel='amplitude ratio',
                     xlim=[0, 22050], 
                     legend=False)
```

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def plot_phase_shift(window):
    signal = thinkdsp.SquareSignal(freq=440)
    wave = signal.make_wave(duration=1, framerate=44100)
    spectrum = wave.make_spectrum()
    ys = numpy.convolve(wave.ys, window, mode='same')
    smooth = thinkdsp.Wave(ys, framerate=wave.framerate)
    spectrum2 = smooth.make_spectrum()
    angles = spectrum.angles()
    angles2 = spectrum2.angles()
    phase_shift = angles2 - angles    
    thinkplot.plot(phase_shift)
    thinkplot.config(xlabel='frequency (Hz)',
                         ylabel='phase shift',
                         xlim=[0, 22050], ylim=[-PI2,PI2],
                         legend=False)
    thinkplot.show()
    
box_car_window = numpy.ones(11)
box_car_window /= sum(box_car_window)
box_car_window = zero_pad(box_car_window, len(wave))
plot_phase_shift(box_car_window)
gaussian = signal.gaussian(M=20000, std=2)
gaussian /= sum(gaussian)
plot_phase_shift(gaussian)
```

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
signal = thinkdsp.ParabolicSignal(freq=440)
wave = signal.make_wave(duration=5*signal.period, framerate=44100)
wave.plot()
thinkplot.show()
spectrum = wave.make_spectrum()
spectrum.differentiate()
spectrum.make_wave().plot()
```

These two graphs make a lot of sense, since the derivative is zero at the minimum of the parabolas, where the slope is zero. Cool, this is the derivative!
