2) Compute the autocorrelation of BitCoin prices. Does the autocorrelation function decay exponentially? Is there evidence of periodic behavior?

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    n = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:n-lag]
    corr = numpy.corrcoef(y1, y2)[0, 1]
    return corr

def autocorr(wave):
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags,corrs = autocorr(wave)
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='Lag',
ylabel='correlation',xscale='log')
```

```python
thinkplot.plot(lags,corrs)
thinkplot.show(xlabel='Lag',
ylabel='correlation')
```

The autocorrelation function does not decay exponentially, since when the function is plotted on a log-x scale, the function is not a straight line. Instead, it looks like the correlation decreases somewhat linearly. There is some evidence of periodic behavior, since there are peaks in correlation about 250 samples apart, at a lag of 200,450, and 700.  

3) Use autocorrelation to estimate the pitch of the saxophone recording in 100475__iluppai__saxophone-weep.wav.  What happens if you filter out the fundamental tone?  Is the estimate based on the autocorrelation function consistent with your perceived pitch? (For my answer, I followed along very closely with Allen's saxophone2.ipynb.

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
segment = wave.segment(start=3.4,duration=0.4)
segment.make_spectrum().plot(high=1000)
thinkplot.show(xlabel='Frequency (Hz)', ylabel='Amplitude')
segment.make_audio()
```

```python
lags,corrs = autocorr(segment)
thinkplot.plot(corrs[:500])
thinkplot.show(xlabel='Lag',ylabel='correlation')

def get_peak_index(low,high,corrs):
    """returns index of the highest value in corrs[low:high]."""
    return numpy.array(corrs[low:high]).argmax() + low
    
max_corr_lag = get_peak_index(80,120,corrs)
period = max_corr_lag / segment.framerate
frequency = 1 / period
print('The percieved pitch is %s Hz'%(frequency))
```

The fundamental frequency is not the dominant frequency! We hear the fundamental frequency at about ~416 Hz, which corresponds to a G#4/Ab4

analyze1, which performs the discrete cosine transform by solving for the amplitude vector with numpy.linalg, looks like it could be a O(n^2) operation or O(n^3) operation because the running time vs. n, when plotted on a log scale, has a slope of 2.2. Given that we already know that solving a linear system of equations should be a O(n^3) operation, I think it's allowable to assume that we just don't have enough experimental data points, and the order of analyze1 is actually (n^3). (Testing with larger n would have taken way too long to run).

```python
ns = 2 ** numpy.arange(4, 14)
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o dct.analyze2(ys, freqs, ts)
    results.append(result)
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
print(t)
```

analyze2 assumes that the M matrix is orthogonal, which we ensured by shifting the frequencies and times by a half. Then analyze2 just takes the inverse of the M matrix and multiplies it by the wave array to get the amplitude array. Since getting the inverse of an orthogonal matrix is almost free, what affects time is the matrix multiplication, which is an O(n^2) operation. The slope of 1.91 of the running time vs. n graph on a log scale matches with what we expect, as it's fairly close to 2. 

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
noise = signal.make_wave(duration=50.0)
ns = 2 ** numpy.arange(4, 22)
results = []
for N in ns:
    ts = (0.5 + numpy.arange(N)) / N
    freqs = (0.5 + numpy.arange(N)) / 2
    ys = noise.ys[:N]
    result = %timeit -o scipy.fftpack.dct(ys)
    results.append(result)
bests = [result.best for result in results]
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
x = numpy.log(ns)
y = numpy.log(bests)
t = scipy.stats.linregress(x,y)
print(t)
```

It looks like scipy's discrete cosine transform is really darn good. I expected it to have O(n^2) running time, but could not turn the knobs enough to see that behavior. The slope of the running time vs. n on a log scale was about .80, though if you cut off the first segment of the wave between 10^1 and 10^3, the slope would probably be around 1. This seems to indicate O(n) running time. Some quick research on scipy's dct II function turned up a paper where an author claimed to have achieved  "17/9N log2N+O(N)" speeds, so I'd buy that scipy's dct II does something really clever to get a nearly linear order of growth. The paper is here: http://arxiv.org/pdf/cs/0703150.pdf

5) One of the major applications of the DCT is compression for both sound and images. In its simplest form, DCT-based compression works like this:
Break a long signal into segments.
Compute the DCT of each segment.
Identify frequency components with amplitudes so low they are inaudible, and remove them. Store only the frequencies and amplitudes that remain.
To play back the signal, load the frequencies and amplitudes for each segment and apply the inverse DCT.
Implement a version of this algorithm and apply it to a recording of music or speech. How many components can you eliminate before the difference is perceptible? *Got help from Joshe, Abi, and Nitya

```python
def cutoff_freqs(hs,cutoff):
    for i, h in enumerate(hs):
        if h < cutoff:
            hs[i] = 0
    return hs

def compress_wave(wave,cutoff):
    segs= []
    for i in range(0,10,1):
        seg_spec = wave.segment(i,duration = 1).make_spectrum()
        seg_spec.hs = cutoff_freqs(seg_spec.hs, cutoff)
        wave_seg = seg_spec.make_wave()
        segs.append(wave_seg)
    compressed_wave = segs[0]
    for i in range(1,len(segs)):
        compressed_wave = compressed_wave | segs[i]
    return compressed_wave

wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
compressed_wave = compress_wave(wave, 1000)
compressed_wave.plot()
thinkplot.show(xlim=[1.0,1.05])
compressed_wave.make_audio()
```

```python
compressed_wave = compress_wave(wave, 100000)
compressed_wave.plot()
thinkplot.show(xlim=[1.0,1.05])
compressed_wave.make_audio()
```

```python
compressed_wave = compress_wave(wave, 10000000)
compressed_wave.plot()
thinkplot.show(xlim=[1.0,1.05])
compressed_wave.make_audio()
```

The difference starts to become very noticable when I cut out frequency components with amplitudes below 10000000. Cool.
