2) Write a class called SawtoothChirp that extends Chirp and overrides evaluate to generate a sawtooth waveform with frequency that increases (or decreases) linearly. Plot the spectrogram of this signal. The effect of aliasing should be visible, and if you listen carefully, you can hear it.


```python
class SawtoothChirp(thinkdsp.Chirp):
    def _evaluate_sawtooth(self, phases):
        mod_phases = numpy.mod(phases,thinkdsp.PI2)
        ys = 2 * thinkdsp.unbias(mod_phases)
        return ys
        
    def _evaluate(self, ts, freqs):
        dts = numpy.diff(ts)
        dphis = thinkdsp.PI2 * freqs * dts
        phases = numpy.cumsum(dphis)
        ys = self.amp * self._evaluate_sawtooth(phases)
        return ys

sawtooth_chirp = SawtoothChirp(start=220,end=880)
sawtooth_chirp_wave = sawtooth_chirp.make_wave(duration = 2)
sawtooth_chirp_wave.make_spectrogram(seg_length=512).plot()
thinkplot.config(title='Sawtooth Chirp Spectrogram',xlabel='Time (s)', ylabel='Amplitude', legend=False)
sawtooth_chirp_wave.make_audio()
```

The effect of aliasing is visible as the lines with negative slope. Once a frequency rises above half of the sampling rate, it "folds over" and this is visible as the line "reflecting" and changing to having a negative slope. Cool! 

3)  In musical terminology, a “glissando” is a note that slides from one pitch to another, so it is similar to a chirp. A trombone player can play a glissando by extending the trombone slide while blowing continuously. As the slide extends, the total length of the tube gets longer, and the resulting pitch is inversely proportional to length.

Assuming that the player moves the slide at a constant speed, how does frequency vary with time? Is a trombone glissando more like a linear or exponential chirp?

Write a function that simulates a trombone glissando from C3 up to F3 and back down to C3. C3 is 262 Hz; F3 is 349 Hz.

```python
class TromboneGlide(thinkdsp.Chirp):
    
    def evaluate(self, ts):
        start = 1. / self.start
        end = 1. /self.end
        freqs = 1. / numpy.linspace(start, end, len(ts)-1)
        return self._evaluate(ts, freqs)
    
def play_trombone(low=262, high=349):
    signal = TromboneGlide(low,high)
    trombone_rise = signal.make_wave(duration = 2)
    signal = TromboneGlide(high,low)
    trombone_drop = signal.make_wave(duration = 2)
    trombone = trombone_rise | trombone_drop
    return trombone
    
wave = play_trombone()
wave.make_audio()

thinkplot.show()
TromboneGlide(262,349).make_wave(duration = .02).plot()
thinkplot.config(title='Trombone',xlabel='time (s)', ylabel='amplitude', legend=False)
thinkplot.show()
thinkdsp.Chirp(262,349).make_wave(duration = .02).plot()
thinkplot.config(title='Linear',xlabel='time (s)', ylabel='amplitude', legend=False)
thinkplot.show()
thinkdsp.ExpoChirp(262,349).make_wave(duration = .02).plot()
thinkplot.config(title='Exponential',xlabel='time (s)', ylabel='amplitude', legend=False)
    
```

The frequency increases like it's traveling to the left along a 1/x curve. A trombone glissando is more like a exponential chirp, from the three graphs above, you can see that it's amplitude is closer to the amplitude of a exponential chirp as compared to the amplitude of the linear chirp after the same amount of time.

6) In a white noise signal, the mixture of frequencies changes over time. In the long run, we expect the power at all frequencies to be equal, but in any sample, the power at each frequency is random.

To estimate the long-term average power at each frequency, we can break a long signal into segments, compute the power spectral density for each segment, and then compute the average across the segments. 

You can read more about this algorithm at http://en.wikipedia.org/wiki/Bartlett's_method.

Apply Bartlett's method to your natural noise source and see if it gives a clearer picture of the relationship between frequency and power.


```python
def get_spec_map(wave, seg_length):
    n = len(wave.ys)
    window = numpy.hamming(seg_length)
    start, end, step = 0, seg_length, seg_length / 2
    spec_map = {}
    while end < n:
        ys = wave.ys[start:end] * window
        hs = numpy.fft.rfft(ys)
        t = (start + end) / 2.0 / wave.framerate
        spec_map[t] = thinkdsp.Spectrum(hs, wave.framerate)
        start += step
        end += step
    return spec_map

def get_avg_power(wave, seg_length):
    spec_map = get_spec_map(wave, seg_length)
    num_segs = len(spec_map.keys())
    random_spectrum = spec_map.values()[0]
    avg_power = [0]*len(random_spectrum.power)
    freqs = random_spectrum.fs
    for val in spec_map.values():
        avg_power += val.power
    avg_power /= num_segs
    thinkplot.plot(freqs,avg_power)

avg_spec = get_avg_power(cricket, 512)
```

As suspected, the cricket signal is not just noise at all. There is a large peak in power at the frequencies of  2500 Hz and 22000 Hz, and this average power spectrum illustrates that very nicely.
